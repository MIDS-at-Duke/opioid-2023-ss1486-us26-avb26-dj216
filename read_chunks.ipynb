```python
state = "FL"


def filter_condition(chunk):
    return chunk[chunk["REPORTER_STATE"] == state]


# empty list to store filtered chunks
filtered_chunks = []

# Read the file in chunks and filter based on the condition
for chunk in pd.read_csv("arcos_all_washpost.tsv", delimiter="\t", chunksize=100000):
    filtered_chunk = filter_condition(chunk)
    filtered_chunks.append(filtered_chunk)

# Concatenate the filtered chunks into a single DataFrame
FL_df = pd.concat(filtered_chunks)
```

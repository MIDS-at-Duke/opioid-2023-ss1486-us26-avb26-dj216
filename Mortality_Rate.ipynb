```python
import pandas as pd
```

```python
deaths = pd.read_csv('US_Vital_Stats_Deaths.csv')
```

```python
population = pd.read_csv('pop_merge.csv')
```

```python
deaths.head(10)
```

```python
deaths[['CTYNAME','STNAME']] = deaths['County'].str.split(',', n=1, expand=True)
```

```python
deaths = deaths[['CTYNAME', 'STNAME','County', 'County Code', 'Year', 'Year Code',
       'Drug/Alcohol Induced Cause', 'Drug/Alcohol Induced Cause Code',
       'Deaths']]
```

```python
deaths.head(10)
```

```python
deaths['Year'] = deaths['Year'].astype(int) 
```

```python
deaths.dtypes
```

```python
stats_abb = pd.read_csv('states_abb.csv')
```

### Merge state_abb with deaths dataset

```python
deaths.columns
```

```python
stats_abb = stats_abb.rename(columns={'Abbreviation': 'STNAME'})
```

```python
deaths['STNAME'] = deaths['STNAME'].str.strip()
```

```python
merge_data = pd.merge(deaths,stats_abb,how='left',on='STNAME')
```

### Merge it with population data 

```python
population.head(10)
```

```python
population = population.rename(columns={'STNAME': 'State'})
```

```python
final_merge = pd.merge(merge_data,population,how='inner',on=['State','CTYNAME','Year'])
```

```python
final_merge['Deaths'] = final_merge['Deaths'].astype(float).astype(int)
```

```python
final_merge.dtypes
```

```python
final_merge['Mortality_Rate'] = final_merge['Deaths'] / final_merge['Population']
```

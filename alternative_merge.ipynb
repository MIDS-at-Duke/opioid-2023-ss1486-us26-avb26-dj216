```python
import pandas as pd
```

```python
db_opiods = pd.read_parquet("county_mme.parquet")
db_mortality = pd.read_csv("Mortality_Rate_Data.csv")
```

```python
db_mortality
```

```python
# pivot drug and alcohol induced cause  morality db
db_mortality = db_mortality.pivot_table(
    index=["State", "CTYNAME", "Year", "Deaths", "Population", "Mortality_Rate"],
    columns="Drug/Alcohol Induced Cause",
    values="Deaths",
).reset_index()
```

```python
# Add state key to mortality dataset
# Create dictionary full state names and their initials
state_mapping = {
    "Alabama": "AL",
    "Alaska": "AK",
    "Arizona": "AZ",
    "Arkansas": "AR",
    "California": "CA",
    "Colorado": "CO",
    "Connecticut": "CT",
    "Delaware": "DE",
    "Florida": "FL",
    "Georgia": "GA",
    "Hawaii": "HI",
    "Idaho": "ID",
    "Illinois": "IL",
    "Indiana": "IN",
    "Iowa": "IA",
    "Kansas": "KS",
    "Kentucky": "KY",
    "Louisiana": "LA",
    "Maine": "ME",
    "Maryland": "MD",
    "Massachusetts": "MA",
    "Michigan": "MI",
    "Minnesota": "MN",
    "Mississippi": "MS",
    "Missouri": "MO",
    "Montana": "MT",
    "Nebraska": "NE",
    "Nevada": "NV",
    "New Hampshire": "NH",
    "New Jersey": "NJ",
    "New Mexico": "NM",
    "New York": "NY",
    "North Carolina": "NC",
    "North Dakota": "ND",
    "Ohio": "OH",
    "Oklahoma": "OK",
    "Oregon": "OR",
    "Pennsylvania": "PA",
    "Rhode Island": "RI",
    "South Carolina": "SC",
    "South Dakota": "SD",
    "Tennessee": "TN",
    "Texas": "TX",
    "Utah": "UT",
    "Vermont": "VT",
    "Virginia": "VA",
    "Washington": "WA",
    "West Virginia": "WV",
    "Wisconsin": "WI",
    "Wyoming": "WY",
}
# Add a new column 'State_Initials' to the second table
db_mortality["State_Initials"] = db_mortality["State"].map(state_mapping)
```

```python
db_mortality["COUNTY_NAME"] = (
    db_mortality["CTYNAME"].str.replace(" County", "", regex=False).str.upper()
)
```

```python
db_mortality["merge_key"] = (
    db_mortality["State_Initials"] + "_" + db_mortality["COUNTY_NAME"]
)
db_opiods["merge_key"] = db_opiods["BUYER_STATE"] + "_" + db_opiods["BUYER_COUNTY"]
```

```python
df_opiods_by_state_county = (
    db_opiods.groupby(["YEAR", "BUYER_STATE"]).count().reset_index()
)
df_mortality_by_state_county = (
    db_mortality.groupby(["Year", "State_Initials"]).count().reset_index()
)
```

```python
df_opiods_by_state_county = df_opiods_by_state_county[
    ["YEAR", "BUYER_STATE", "BUYER_COUNTY"]
]
df_opiods_by_state_county = df_opiods_by_state_county.rename(
    columns={"BUYER_COUNTY": "number_counties_opiod_df"}
)
df_opiods_by_state_county
```

```python
df_mortality_by_state_county = df_mortality_by_state_county[
    ["Year", "State_Initials", "CTYNAME"]
]
df_mortality_by_state_county = df_mortality_by_state_county.rename(
    columns={"CTYNAME": "number_counties_mortality_df"}
)
```

```python
df_mortality_by_state_county
```

```python
# Assuming df1 is the first DataFrame and df2 is the second DataFrame with State Initials column
merged_df = pd.merge(
    df_opiods_by_state_county,
    df_mortality_by_state_county,
    left_on=["YEAR", "BUYER_STATE"],
    right_on=["Year", "State_Initials"],
    how="left",
)
```
